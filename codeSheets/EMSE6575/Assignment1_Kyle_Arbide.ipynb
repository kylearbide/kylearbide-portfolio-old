Classification: Using a model to apply a discrete label given data. These types of models are considered supervised learning
models. An example would be training a KNeighbors model to idenify the animal in a photograph using a labeled dataset 
of different animals.

Clustering: Using a model to identify groups of datapoints that act similarly. Clustering models are unsupervised 
learning models. An example of this is to use KMeans clustering to understanding the purchasing habits of different kinds of
shoppers at a grocery store.

Density Estimation: Estimation of the probability density function using a sample of data. An example could be given a dataset of the heights of 100 people, use Kernel Density Estimation to estimate the probability of being 6'2".

Regression: Using a set of explanatory variables to predict a dependent variable. It can predict both continuous 
values (Linear Regression) and discrete values (Logistic Regression). Regression models are supervised learning models. An 
example of this is using linear regression to find a formula that predicts a person's height based on their weight, gender, etc.

Samples: A sample is a random subset of data taken from a population. The sample data is used as a representation of the population and its qualities. An example of this is if we take the data of all the actions of 1000 twitter users. The population of twitter users is much larger but it is much easier to work with a sample. 

Features: A feature is a descriptive variable to provides information about a data point. In regression, our independent variables are also known as our features. In the Height Regression example given above, some features might include weight, sex, age, etc. 

Multivariate Data: Multivariate data contains multiple descriptive variables for each data point. The Height regression example we've mentioned uses multivariate data because there are multiple descriptors (weight, sex,age,etc.)

Training Set: When we split our sample data, the training set will be used to train our machine learning model. When we split the data, this will usually be the largest set we create. If we have a sample of 100 people for our height regression model, we might take 70 of these samples to train our model.

Test Set: The test set is used to test the accuracy of our trained model. It is important that this set is seperated from training to ensure the model has never seen this data before. In the example given above the remaining 30 samples would be used as the test set. 

Holdout Set: Another name for the test set, the set that isn't seen by the model during training. See example from above.

```python
#Question 2 part 2
import matplotlib.pyplot as plt
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split #Here we are importing packages like earlier, but this time we also include
                                                     #The meterics and train_test_split packages from sklearn and matplotlib
```

```python
digits = datasets.load_digits() # Load in the digits dataset

_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3)) # create 4 subplots of size 10,3
for ax, image, label in zip(axes, digits.images, digits.target): #for each of the first 4 images in the digits dataset
    ax.set_axis_off() #turn of the x and y axis
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest") #show the image from the dataset
    ax.set_title("Training: %i" % label)    #Label with the digit that is being displayed
```

```python
n_samples = len(digits.images)     #Take the number of samples in the digits dataset
data = digits.images.reshape((n_samples, -1)) #Reshape the images into (64,)

clf = svm.SVC(gamma=0.001) #create our svm classifies with specifications like before

X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.5, shuffle=False) #split the new/re-shaped data into training and testing sets,
                                                       #we specify a 50/50 split

clf.fit(X_train, y_train) # train the model using the features and responses from the training set

predicted = clf.predict(X_test) # use the trained model to predict the testing set
```

```python
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction in zip(axes, X_test, predicted):
    ax.set_axis_off()
    image = image.reshape(8, 8) #We need to return the image to its original shape
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Prediction: {prediction}") #This is the same as the plotting above, except here we are plotting the predicted
                                              #values with their corresponding features
```

```python
print(
    f"Classification report for classifier {clf}:\n"
    f"{metrics.classification_report(y_test, predicted)}\n" # this uses the true responses and the predicted responses to show
                                                            # the different accuracy scores for each predicted digit
) 
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, predicted) #The predictions and true results displayed in a confusion matrix
disp.figure_.suptitle("Confusion Matrix") #plot title
print(f"Confusion matrix:\n{disp.confusion_matrix}") # the confusion matrix but printed instead of plotted

plt.show()
#Aparently this doesnt work anymore so lets try another method
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=clf.classes_)
disp.plot()
plt.show()
print(f"Confusion matrix:\n{cm}")
```

```python
#Question 3 
from sklearn import datasets, svm
from sklearn.model_selection import train_test_split 
iris = datasets.load_iris()
print(iris.data[:5])
print(iris.target[:5])
print(iris.feature_names)
print(iris.target_names)
# We load the iris dataset and show the first 5 feature vectors and targets. We see what our feature
# columns describe, as well as our different classification labels
```

```python
clf = svm.SVC(gamma=0.001)

X_train,X_test, y_train, y_test = train_test_split(
    iris.data, iris.target_names[iris.target], test_size=25, shuffle=True) 

clf.fit(X_train,y_train)

predicted = clf.predict(X_test)
# We load the support vector classifier, split the data into training and testing sets, fit the model
# to the training sets and create predictions using the testing features
```

```python
print(y_test)
print(predicted)
# We display the predicted values against the actual target values
```

```python
def top_bigrams(text, n):
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    finder = nltk.collocations.BigramCollocationFinder.from_words(text)
    return(finder.nbest(bigram_measures.pmi, n))
```
